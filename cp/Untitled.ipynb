```python
import scipy.optimize
import numpy as np
from scipy.optimize import fsolve
```

```python
def beta_equation(beta):
    y = 3*beta - np.tan(beta*3)
    return y
```

```python
points = np.arange(-0.1, 1e3, 1e-4)
betas = []

for beta in points:
    if beta_equation(beta)>0:
        flag = 0
    elif beta_equation(beta)<0 and flag==0:
        betas.append(beta)
        flag = 1
print(betas)
```
